```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/CodeYouOrg/DataOpenClass/refs/heads/main/SalaryData.csv'
df = pd.read_csv(url)

# dropping columns we dont need
columns_to_drop = ['Overtime_Rate',	'Incentive_Allowance', 'Other', 'ObjectId']
df = df.drop(columns=columns_to_drop)

# We dont have a ton of missing data so we are dropping the NaNs 
df = df.dropna()

df = df[(df['CalYear'] == 2022) & (df['YTD_Total'] >= 100000)]

top_10_earners = df.groupby('Employee_Name')['YTD_Total'].sum().sort_values(ascending=False).head(10)
top_10_earners.plot(
    kind='bar',
    xlabel = 'Employee_Name',
    ylabel = 'YTD_Total',
    title='Top 10 Salaries in 2022',
    fontsize='18'
    )

print(top_10_earners.plot)
```
